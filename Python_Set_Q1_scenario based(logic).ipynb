```python
# 2. Scenario: A program processes a list of numbers and needs to find the largest value.
def largest_value():
    numbers = [25, 35, 6, 18, 5, 2, 129]
    largest = numbers[0]
    for i in numbers:
        if(i>largest):
            largest = i
            
    print(largest, "is the largest value in list")
    return largest
       
```

```python
largest_value()
```

```python
# 3.Scenario: A company provides employees with a 10% bonus if their salary exceeds $50,000.
def Bonus():
    salary = [40000, 50000, 60000]
    for sal in salary:
        if(sal>50000):
            bonus = sal*0.10
            tot_sal = "Total Salary:", bonus + sal
    
           
            print("Salary:$",sal, "Bonus:$",bonus)
            
        else:
            print("Salary:$",sal, "No bonus")
            
    return tot_sal
            
```

```python
Bonus()
```

```python
# 4. Scenario: A program evaluates a number to determine if it is even or odd.
def Evenorodd():
    number=int(input("Enter any number"))
    if(number%2==0):
        print(number, "is Even Number")
        result = "Even number", number
    else:
        print(number,"is Odd Number")
        result = "Odd Number", number
    return result

```

```python
# 8. Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
def Account_Balance():
    money = int(input("Enter the amount to Withdrawal:"))
    avail_bal = 100000
    if(money<=avail_bal):
        new_balance = avail_bal-money
        print(new_balance)
        slip = "Available Balance = ", new_balance
    else:
        slip = "No Sufficient Balance"
    return slip
                
```

```python
# 9. Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules.
def Find_LeapYear():
    year = int(input("Enter the year to check leap year:"))
    if(year%400==0):
        LP = year, "is a Leap Year"
    elif(year%100==0):
        LP = year, "is a not Leap Year"
    
    elif(year%4==0):
        LP = year, "is a Leap Year" 
    else:
        LP = year, "is not a Leap Year"
    return LP
            
               
```

```python
Find_LeapYear()
```

```python
# 10. Scenario: A program filt'ers out only even numbers from a given list.
def Filter_EvenNumbers():
    data_list = [28, 63, 33, 74, 66, 99]
    even_numbers = [number for number in data_list if number % 2 == 0]
    return even_numbers
       
```
